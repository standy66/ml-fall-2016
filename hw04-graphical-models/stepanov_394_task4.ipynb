**Вопрос 1**: Зачем нужно структурное предсказание?

Чтобы предсказывать метки (или действительные значения) для объектов, которые не являются независимыми (либо метки которых не являются независимыми)

**Вопрос 2**: Что такое сопряженное распределение?

Если $p(\theta)$ принадлежит тому жу семейству распределений, что и $p(\theta | x)$, то $p(\theta)$ и $p(x | \theta)$ называются сопряженными.

**Вопрос 3**: Какое распределение сопряженное к равномерному, докажите?

Распределение Парето.

Доказательство:


Пусть $p(x | \theta) \sim U[0, \theta]$, $p(\theta)$ -- распределение Парето с параметрами $\theta_0$ и $k$, т.е. 
$$p(\theta) = \begin{cases}
\frac{k\theta_0^k}{\theta^{k+1}} ,& \theta \geq \theta_0 \\
0,& \theta < \theta_0 
\end{cases}$$.
Тогда $p(\theta | x)$ -- тоже распределение Парето с параметрами $\theta_0' = \max \{\theta_0, x\}, k' = k + 1$. Действительно, $p(\theta | x) \propto p(x | \theta) p(\theta)$. Если $\theta < \theta_0$, то $p(\theta) = 0$, значит и $p(\theta | x)$. Если $\theta < x$, то $p(x | \theta) = 0$, т.к. оно равномерное на отрезке $[0, \theta]$. То есть при $\theta < \theta_0'$ выполнено, что $p(\theta | x) = 0$, если же $\theta > \theta_0'$, то $p(\theta | x ) \propto \frac{1}{\theta} \frac{1}{\theta^{k+1}}$, то есть $p(\theta | x)$ -- распределение Парето с параметрами $\theta_0', k'$


Доказано.

**Вопрос 4**: В чем заключается Байесовский подход к машинному обучению?

Предполагаем, что есть совместное "распределение" $p(X, Y, W)$, где $X$ -- объекты, $Y$ -- метки, $W$ -- веса модели, $p$ -- баесовская вероятность, "степень уверенности в значениях X, Y, W". Далее оптимальные веса модели находятся из условия $W^* = \arg \max \limits_{W} p(X, Y, W)$, а классификация осуществляется путем нахождения $Y^* = \arg \max \limits_{Y} p(X, Y, W^*)$

**Вопрос 5**: В чем основные преимущества Байесовского подхода к машинному обучения?

Можно задавать наши предпочтения на параметры $W$ основываясь на экспертном знании. В результате использования Баесовских методов в машинном обучении, мы получаем не просто метки классов, а распределения по классам (или фичам).

**Вопрос 6**: Чем отличается структурный метод опорных векторов от неструктурного, в чем сложности при обучении структурного?

Структурный метод опорных векторов применяется для структурного предсказания, где объекты и ответы на них зависимы, неструктурный -- для обычного предсказания, когда объекты независимы. В неструктурном методе максимизируется ширина зазора, а в структурном -- минимизируется энергия. Сложности структурного SVM заключаются в том, что неравенств-ограничений экспоненциальное число, и даже вывод (inference) не всегда удается провести точно, часто используют приближенные методы.

#### 1. Введите граф модель, почему вы выбрали именно такую, приложите рисунок

Введем сетку в качестве графической модели. В ней каждая вершина соответствует пикселю, каждая вершина зависит от 4 соседних. Это позволяет моделировать тот факт, что соседние пиксели с большей вероятностью могут иметь одну и ту же метку, чем не соседние.

#### 2. Потенциалы каких порядков вы собираетесь использовать?

Потенциалы второго порядка: унарные и парные. Их предлагает использовать Дмитрий Ветров (https://www.youtube.com/watch?v=VOOIxnXA26Q).

В этом случае энергия переписывается таким образом: 

$$E(X) = \sum_{i} \varphi_i(x_i) + \sum_{i, j} \varphi_{i, j}(x_i, x_j)$$, где $\varphi_i$ -- унарные потенциалы, $\varphi_{i, j}$ -- парные, $X = (x_1, \dots, x_n)$ -- сегментация изображения, $x_i \in \{1, 2, 3, 4\}$ -- номер изображения, которому должен принадлежать $i$-тый пиксель.

#### 3. Определите потенциалы -- какой в них физический смысл? почему они поощряют незаметные склеивания?

$\varphi_i(x_i) = I\{x_i \in D_i\}$, где $D_i$ -- множество номеров изобрежений, где $i$-тый пиксель не принадлежит человеку, либо пустое, если ни на какой картинке человека нет. В предположении, что человек для каждого пикселя есть только на одной картинке, можно в качетсве номеров изображений, которые помещаются в $D_i$ взять топ 3 пикселей, ближайщих в среднему из 4.

Либо можно попросить человека выбрать, из какой картинки он хочет взять ту или иную область (таким образом человек задаст унарные потенциалы), а парные потенциалы позволят склеить изображения без швов.

$\varphi_{i, j}(x_i, x_j) = I\{x_i \neq x_j\} + \nabla_{i, j}(x_i, x_j)$, где $\nabla_{i, j}(x_i, x_j)$ -- модуль градиента между $i$-тым пикселем с изображения $x_i$ и $j$-тым с изображения $x_j$.

Физический смысл потенциалов -- энергия состояния. Первое слагаемое парных потенциалов поощрает короткие склейки, второе -- незаметные переходы.

#### 4. Предложить несколько вариантов выбора потенциалов.

$\varphi_{i, j}(x_i, x_j) = I\{x_i \neq x_j\} \big(1 + \nabla_{i, j}(x_i, x_j) \big)$ -- в этом случае если разреза нет, то потенциал будет ноль, а если есть то как-минимум 1 + еще может быть градиент цвета.

Также можно заменить пиксели на блоки пикселей, а $\nabla_{i, j}$ заменить на норму разности значений цветов для $i$ и $j$ блока.

Можно сделать просто $\varphi_{i, j}(x_i, x_j) = \nabla_{i, j}(x_i, x_j)$, но в этом случае разрезы могут становится очень длинными.

Можно изменить сетку и сделать ее более связной, тогда величину потенциала можно связать с растсоянием между пикселями.
